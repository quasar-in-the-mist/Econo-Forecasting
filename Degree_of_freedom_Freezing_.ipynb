The Key Physics:
The frozen DOFs don't vanish — they transmute into effective operators:

Frozen antiparticle DOFs → spin-orbit, Darwin term, mass corrections
Frozen N-1 particles → effective mass, screened potential, exchange-correlation
This is a general principle: eliminating DOFs always costs you operator complexity.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# =============================================================================
# Physical constants (natural-ish units, but keep c explicit for clarity)
# =============================================================================
c = 1.0        # speed of light
m = 1.0        # rest mass of the charge carrier

# =============================================================================
# 1. DIRAC SPINOR COMPONENT WEIGHTS vs BOOST
#
# For a free Dirac particle with momentum p, the positive-energy spinor is:
#
#   u(p) = N * ( chi,  (sigma.p)/(E+m) * chi )^T
#
# where chi is a 2-component Pauli spinor. The "large" (upper) and "small"
# (lower) components have relative weight:
#
#   |upper|^2 / |total|^2  =  (E+m) / (2E)
#   |lower|^2 / |total|^2  =  (E-m) / (2E)  =  p^2 / [2E(E+m)]
#
# The lower components ARE the antiparticle DOFs that get frozen in FW.
# =============================================================================

def gamma_from_beta(beta):
    return 1.0 / np.sqrt(1.0 - beta**2)

def energy_from_beta(beta, m=1.0):
    return gamma_from_beta(beta) * m * c**2

def momentum_from_beta(beta, m=1.0):
    gamma = gamma_from_beta(beta)
    return gamma * m * beta * c

def upper_weight(beta):
    """Fraction of Dirac spinor in 'large' (particle) components"""
    E = energy_from_beta(beta)
    return (E + m*c**2) / (2.0 * E)

def lower_weight(beta):
    """Fraction of Dirac spinor in 'small' (antiparticle) components"""
    E = energy_from_beta(beta)
    return (E - m*c**2) / (2.0 * E)

# =============================================================================
# 2. EFFECTIVE DOF COUNT
#
# We model the effective DOFs as:
#   - 3 translational (always active)
#   - 2 spin DOFs (always active for spin-1/2)
#   - 2 "antiparticle" spinor DOFs (activate with boost)
#   - Additional pair-creation channels (activate at high energy)
#
# The effective number of internal DOFs:
#   n_eff = 2 (spin) + 2 * f_lower(gamma)  [spinor DOFs]
#          + n_pair * f_pair(gamma)          [pair channels]
# =============================================================================

def spinor_dofs(beta):
    """Effective number of spinor DOFs (2 to 4)"""
    return 2.0 + 2.0 * lower_weight(beta) / 0.5  # normalized so it→4 as beta→1

def pair_channel_dof(beta, threshold_gamma=2.0, width=0.5):
    """
    Pair creation channels open when gamma*mc^2 > 2mc^2
    Modeled as a smooth activation (Fermi function)
    """
    gamma = gamma_from_beta(beta)
    # Activation function: sigmoid around threshold
    x = (gamma - threshold_gamma) / width
    return 1.0 / (1.0 + np.exp(-x))

def total_effective_dofs(beta, n_pair_channels=3):
    """
    Total effective DOFs:
    - 3 spatial (translation) -- always on
    - spinor DOFs: 2 (NR) to 4 (ultrarelativistic)
    - pair channels: 0 (NR) to n_pair_channels (ultrarelativistic)
    """
    spatial = 3.0
    spinor = spinor_dofs(beta)
    pairs = n_pair_channels * pair_channel_dof(beta)
    return spatial, spinor, pairs

# =============================================================================
# 3. FOLDY-WOUTHUYSEN "DECOUPLING" VISUALIZATION
#
# The FW transformation effectiveness can be measured by the ratio
# of the off-diagonal (coupling) terms to diagonal terms in the
# Dirac Hamiltonian. In the standard representation:
#
#   H_Dirac = ( mc^2 + V    c*sigma.p )
#             ( c*sigma.p   -mc^2 + V  )
#
# The "coupling strength" ~ |c*p| / |mc^2| = beta*gamma = p/(mc)
# FW works when this is small.
# =============================================================================

def fw_coupling_strength(beta):
    """Ratio of off-diagonal to diagonal in Dirac Hamiltonian"""
    gamma = gamma_from_beta(beta)
    return beta * gamma  # = p/(mc)

def fw_decoupling_quality(beta):
    """How well FW decouples particle/antiparticle sectors (1=perfect, 0=none)"""
    coupling = fw_coupling_strength(beta)
    return 1.0 / (1.0 + coupling**2)

# =============================================================================
# 4. MANY-BODY SELF-ENERGY MODEL
#
# For the N→1 particle reduction, model a simple self-energy
# that shows how the quasiparticle residue Z changes.
# Z=1: perfect quasiparticle (frozen background is inert)
# Z<1: DOFs leaking into many-body excitations
# =============================================================================

def quasiparticle_residue(beta, coupling_g=0.3):
    """
    Simple model: Z = 1/(1 + g^2 * gamma)
    Higher boost → more phase space for many-body excitations → smaller Z
    """
    gamma = gamma_from_beta(beta)
    return 1.0 / (1.0 + coupling_g**2 * (gamma - 1.0))

def effective_mass_ratio(beta, coupling_g=0.3):
    """m*/m from self-energy: increases with coupling to frozen DOFs"""
    Z = quasiparticle_residue(beta, coupling_g)
    return 1.0 / Z

# =============================================================================
# PLOTTING
# =============================================================================

beta_values = np.linspace(0.001, 0.999, 1000)
gamma_values = gamma_from_beta(beta_values)

fig = plt.figure(figsize=(16, 20))
gs = GridSpec(4, 2, hspace=0.35, wspace=0.3)

# --- Plot 1: Dirac Spinor Component Weights ---
ax1 = fig.add_subplot(gs[0, 0])
upper = np.array([upper_weight(b) for b in beta_values])
lower_comp = np.array([lower_weight(b) for b in beta_values])

ax1.fill_between(gamma_values, 0, upper, alpha=0.4, color='steelblue', label='Upper (particle)')
ax1.fill_between(gamma_values, upper, upper + lower_comp, alpha=0.4, color='coral', label='Lower (antiparticle)')
ax1.set_xscale('log')
ax1.set_xlabel(r'$\gamma = (1-\beta^2)^{-1/2}$', fontsize=12)
ax1.set_ylabel('Spinor component weight', fontsize=12)
ax1.set_title('Dirac Spinor Decomposition', fontsize=13, fontweight='bold')
ax1.legend(fontsize=11)
ax1.set_ylim(0, 1.05)
ax1.axvline(x=1.1, color='green', linestyle='--', alpha=0.5, label=r'NR regime')
ax1.axvline(x=10, color='red', linestyle='--', alpha=0.5, label=r'Ultrarelativistic')

# --- Plot 2: Effective DOF Count ---
ax2 = fig.add_subplot(gs[0, 1])
spatial_dofs = []
spinor_dofs_list = []
pair_dofs_list = []

for b in beta_values:
    s, sp, p = total_effective_dofs(b)
    spatial_dofs.append(s)
    spinor_dofs_list.append(sp)
    pair_dofs_list.append(p)

spatial_dofs = np.array(spatial_dofs)
spinor_dofs_arr = np.array(spinor_dofs_list)
pair_dofs_arr = np.array(pair_dofs_list)
total_dofs = spatial_dofs + spinor_dofs_arr + pair_dofs_arr

ax2.fill_between(gamma_values, 0, spatial_dofs, alpha=0.4, color='gold', label='Spatial (3)')
ax2.fill_between(gamma_values, spatial_dofs, spatial_dofs + spinor_dofs_arr,
                  alpha=0.4, color='steelblue', label='Spinor (2→4)')
ax2.fill_between(gamma_values, spatial_dofs + spinor_dofs_arr, total_dofs,
                  alpha=0.4, color='coral', label='Pair channels')
ax2.plot(gamma_values, total_dofs, 'k-', linewidth=2, label='Total')
ax2.set_xscale('log')
ax2.set_xlabel(r'$\gamma$', fontsize=12)
ax2.set_ylabel('Effective DOFs', fontsize=12)
ax2.set_title('DOF Activation vs Boost', fontsize=13, fontweight='bold')
ax2.legend(fontsize=10)
ax2.axhline(y=5, color='green', linestyle=':', alpha=0.7, label='NR: 3+2=5')

# --- Plot 3: FW Decoupling Quality ---
ax3 = fig.add_subplot(gs[1, 0])
fw_quality = np.array([fw_decoupling_quality(b) for b in beta_values])
fw_coupling = np.array([fw_coupling_strength(b) for b in beta_values])

ax3.plot(gamma_values, fw_quality, 'b-', linewidth=2.5, label='Decoupling quality')
ax3.fill_between(gamma_values, fw_quality, alpha=0.2, color='steelblue')
ax3.set_xscale('log')
ax3.set_xlabel(r'$\gamma$', fontsize=12)
ax3.set_ylabel('FW Decoupling Quality', fontsize=12)
ax3.set_title('Foldy-Wouthuysen Effectiveness\n(1 = perfect particle/antiparticle separation)',
              fontsize=13, fontweight='bold')
ax3.set_ylim(0, 1.05)
ax3.axhspan(0.9, 1.0, alpha=0.1, color='green', label='Good NR approx.')
ax3.legend(fontsize=11)

# secondary axis: coupling strength
ax3b = ax3.twinx()
ax3b.plot(gamma_values, fw_coupling, 'r--', linewidth=1.5, alpha=0.7, label=r'$p/mc = \beta\gamma$')
ax3b.set_ylabel(r'Coupling strength $\beta\gamma$', color='red', fontsize=12)
ax3b.tick_params(axis='y', labelcolor='red')
ax3b.set_yscale('log')
ax3b.legend(loc='center right', fontsize=10)

# --- Plot 4: Quasiparticle Properties (N→1 reduction) ---
ax4 = fig.add_subplot(gs[1, 1])
Z_values = np.array([quasiparticle_residue(b) for b in beta_values])
mstar_values = np.array([effective_mass_ratio(b) for b in beta_values])

ax4.plot(gamma_values, Z_values, 'b-', linewidth=2.5, label='Quasiparticle residue Z')
ax4.plot(gamma_values, 1.0/mstar_values, 'r--', linewidth=2, label=r'$m/m^*$', alpha=0.8)
ax4.set_xscale('log')
ax4.set_xlabel(r'$\gamma$', fontsize=12)
ax4.set_ylabel('Quasiparticle weight', fontsize=12)
ax4.set_title('N→1 Body Reduction Quality\n(Quasiparticle Picture)', fontsize=13, fontweight='bold')
ax4.legend(fontsize=11)
ax4.set_ylim(0, 1.05)
ax4.axhspan(0.8, 1.0, alpha=0.1, color='green')

# --- Plot 5: Combined "Phase Diagram" of DOF Freezing ---
ax5 = fig.add_subplot(gs[2, :])

# Compute a combined "effective theory quality" metric
combined_quality = fw_quality * Z_values

ax5.plot(gamma_values, fw_quality, 'b-', linewidth=2, label='Rel→NR (FW decoupling)')
ax5.plot(gamma_values, Z_values, 'r-', linewidth=2, label='N-body→1-body (quasi-Z)')
ax5.plot(gamma_values, combined_quality, 'k-', linewidth=3, label='Combined quality')
ax5.fill_between(gamma_values, combined_quality, alpha=0.15, color='gray')

# Mark regimes
ax5.axvspan(1.0, 1.1, alpha=0.2, color='green')
ax5.axvspan(3, 1000, alpha=0.1, color='red')
ax5.text(1.03, 0.5, 'NR\nregime', fontsize=11, ha='center', fontweight='bold', color='green')
ax5.text(30, 0.5, 'Relativistic\nregime', fontsize=11, ha='center', fontweight='bold', color='red')

ax5.set_xscale('log')
ax5.set_xlabel(r'$\gamma$ (boost factor)', fontsize=13)
ax5.set_ylabel('Effective Theory Quality', fontsize=13)
ax5.set_title('DOF Freezing: When Does the 1-Particle NR Description Work?',
              fontsize=14, fontweight='bold')
ax5.legend(fontsize=11, loc='center right')
ax5.set_ylim(0, 1.05)

# --- Plot 6: Energy-level diagram showing DOF freezing ---
ax6 = fig.add_subplot(gs[3, 0])

# Schematic: show how the "spectrum" of a Dirac particle looks
# at different gamma values
gamma_snapshots = [1.01, 1.1, 1.5, 3.0, 10.0]
x_positions = np.arange(len(gamma_snapshots))

for i, g in enumerate(gamma_snapshots):
    beta_i = np.sqrt(1.0 - 1.0/g**2)
    E = g * m * c**2

    # Particle sector energy levels
    E_upper = E
    # Antiparticle sector (negative energy in Dirac sea picture)
    E_lower = -E
    # Gap
    gap = 2 * m * c**2

    # FW-transformed: particle energy
    E_nr = m*c**2 + 0.5 * m * (beta_i * c)**2  # non-relativistic approx

    # Draw levels
    lw = 2.5
    ax6.plot([i-0.3, i+0.3], [E_upper/m, E_upper/m], 'b-', linewidth=lw)
    ax6.plot([i-0.3, i+0.3], [E_lower/m, E_lower/m], 'r-', linewidth=lw, alpha=0.3 + 0.7*(1-fw_decoupling_quality(beta_i)))
    ax6.plot([i-0.2, i+0.2], [E_nr/m, E_nr/m], 'g--', linewidth=lw, alpha=fw_decoupling_quality(beta_i))

    # Draw gap
    ax6.annotate('', xy=(i+0.35, E_lower/m), xytext=(i+0.35, E_upper/m),
                arrowprops=dict(arrowstyle='<->', color='gray', lw=1))

ax6.set_xticks(x_positions)
ax6.set_xticklabels([f'γ={g}' for g in gamma_snapshots], fontsize=10)
ax6.set_ylabel(r'$E / mc^2$', fontsize=12)
ax6.set_title('Energy Levels vs Boost\n(blue=particle, red=antiparticle, green=NR approx)',
              fontsize=12, fontweight='bold')

# --- Plot 7: Partition function / thermodynamic DOF analogy ---
ax7 = fig.add_subplot(gs[3, 1])

# Treat γmc² as an effective "temperature" and compute
# the partition function over internal states
def partition_function(gamma_val):
    """
    Z = sum_i exp(-E_i / kT_eff) where kT_eff ~ gamma * mc^2

    States:
    - 2 spin states: E=0 (always active)
    - 2 antiparticle spinor components: E ~ 2mc^2
    - pair states: E ~ 2mc^2 * n
    """
    kT = gamma_val * m * c**2  # effective energy scale

    Z = 2.0  # spin states (massless internal DOF)
    Z += 2.0 * np.exp(-2.0 * m * c**2 / kT)  # antiparticle components
    # Pair channels
    for n in range(1, 4):
        Z += 2.0 * np.exp(-2.0 * n * m * c**2 / kT)  # e+e- pairs
    return Z

Z_partition = np.array([partition_function(g) for g in gamma_values])

ax7.plot(gamma_values, Z_partition, 'purple', linewidth=2.5)
ax7.fill_between(gamma_values, 2, Z_partition, alpha=0.2, color='purple')
ax7.axhline(y=2, color='green', linestyle='--', alpha=0.7, label='NR limit: Z=2 (spin only)')
ax7.set_xscale('log')
ax7.set_xlabel(r'$\gamma$', fontsize=12)
ax7.set_ylabel('Partition function Z', fontsize=12)
ax7.set_title('Thermal Analogy: DOFs as Partition Function\n' +
              r'($\gamma mc^2$ as effective energy scale)',
              fontsize=12, fontweight='bold')
ax7.legend(fontsize=11)

plt.suptitle('Freezing Degrees of Freedom:\nRelativistic N-Body → Non-Relativistic 1-Particle',
             fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('dof_freezing.png', dpi=150, bbox_inches='tight')
plt.show()

# =============================================================================
# NUMERICAL SUMMARY TABLE
# =============================================================================
print("\n" + "="*80)
print("SUMMARY: Effective DOFs at Various Boost Factors")
print("="*80)
print(f"{'γ':>8} {'β':>8} {'Spinor':>10} {'FW qual':>10} {'Z_qp':>10} {'Total DOF':>12} {'Theory'}")
print("-"*80)

for gamma_target in [1.001, 1.01, 1.1, 1.5, 2.0, 5.0, 10.0, 100.0]:
    beta_t = np.sqrt(1.0 - 1.0/gamma_target**2)
    sp_d = spinor_dofs(beta_t)
    fw_q = fw_decoupling_quality(beta_t)
    z_qp = quasiparticle_residue(beta_t)
    s, sp, p = total_effective_dofs(beta_t)
    tot = s + sp + p

    if fw_q > 0.95 and z_qp > 0.95:
        theory = "NR Schrödinger ✓"
    elif fw_q > 0.8:
        theory = "Pauli + corrections"
    elif fw_q > 0.5:
        theory = "Need Dirac"
    else:
        theory = "Full QFT needed"

    print(f"{gamma_target:>8.3f} {beta_t:>8.4f} {sp_d:>10.3f} {fw_q:>10.4f} {z_qp:>10.4f} {tot:>12.2f} {theory}")

print("\n" + "="*80)
print("KEY INSIGHT:")
print("="*80)
print("""
As γ → 1 (non-relativistic limit):
  • Lower Dirac components → 0     (antiparticle DOFs freeze)
  • FW decoupling → perfect         (particle number conserved)
  • Quasiparticle residue Z → 1     (1-particle picture valid)
  • Effective DOFs: 4 spinor → 2 spin, pair channels → 0

  Total: ~10 DOFs (relativistic) → 5 DOFs (3 spatial + 2 spin)

The 'frozen' DOFs don't disappear — they generate:
  • Mass renormalization (m → m*)
  • Effective potentials (Darwin, spin-orbit, etc.)
  • Higher-order correction operators in 1/c expansion
""")
```
